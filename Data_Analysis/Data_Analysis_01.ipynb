```python
# Trying complex queries

newdf.loc[(newdf[0]<0.3)]     
# All rows will be returned where nos are less than 
# 0.3 in the 0th column
```

```python
newdf.loc[(newdf[0]<0.3) & (newdf[2]>0.1)] 
# All rows will be returned where nos are less than 
# 0.3 in the 0th column and greater than 0.1 in the second column
```

```python
newdf.head(2)
```

```python
newdf.iloc[0,4]         # This will return the exact integer in 0,4 by position 
```

```python
# Difference between loc and iloc

# -Loc is used for label based slicing
# -Iloc is used for position based slicing

# -Index could be str or int but it works only based on labels(loc)
# -Index could be str or int but it works only based on positions(iloc)

# -If label is not present, gives key error(loc)
# -If position is not present, gives index error(iloc)

# -loc can use boolean series as parameter(loc)
# -iloc can't use boolean series as parameter(iloc)
```
